```python
def kmerCommon(dict1,dict2):
    my_dict = [dict1,dict2]

    common_keys = set.intersection(*map(set,my_dict))
    return len(common_keys)
    
```

```python
read={}
```

```python
def globalValue(df):
    
    sequences = eval(df['kmer7'])
    length = len(sequences)
    b = np.zeros(shape=(length,length))
    name  = str(df.target_gene)
    for i in range(0,length):
        for j in range(0,length):
            if(b[j][i] != 0):
                b[i][j] = b[j][i]
            else:
                b[i][j] = kmerCommon(sequences[i],sequences[j])
    read[name] = b
    return b
```

```python
for i in range(0,len(kmer7)):
    globalValue(kmer7.iloc[i])
```

```python
np.save('my_dict_kmer7.npy',  read, allow_pickle=True) 
```
